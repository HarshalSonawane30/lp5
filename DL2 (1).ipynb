```python
"""Binary classification using Deep Neural Networks Example: Classify movie
reviews into positive" reviews and "negative" reviews, just based on the text content of the reviews.Use
IMDB dataset"""
```

```python

# 1. Import Libraries

import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D
```

```python

# 2. Load Dataset

vocab_size = 10000   # top 10k words
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=vocab_size)

print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))
```

```python
# 3. Preprocessing (Padding)

max_length = 200

X_train = pad_sequences(X_train, maxlen=max_length, padding='post')
X_test = pad_sequences(X_test, maxlen=max_length, padding='post')
```

```python
# 4. Build Deep Neural Network

model = Sequential()

# Embedding layer
model.add(Embedding(input_dim=vocab_size, output_dim=128, input_length=max_length))

# Pooling (better than flatten)
model.add(GlobalAveragePooling1D())

# Hidden layers
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))

# Output layer (Binary classification)
model.add(Dense(1, activation='sigmoid'))
```

```python
# 5. Compile Model

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Model Summary
model.summary()
```

```python
# 6. Train Model

history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.2
)
```

```python
# 7. Evaluate Model

loss, accuracy = model.evaluate(X_test, y_test)

print("\nTest Accuracy:", accuracy)
```

```python
# 8. Predict Example

# Predict first test review
prediction = model.predict(X_test[0].reshape(1, -1))

print("\nPrediction Value:", prediction[0][0])

if prediction[0][0] > 0.5:
    print("Review Sentiment: Positive ")
else:
    print("Review Sentiment: Negative ")
```
